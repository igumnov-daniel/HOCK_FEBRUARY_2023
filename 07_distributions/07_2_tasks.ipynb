> Звёзды лучше видны с крыши, полезай и проверь сам. Ты так ждал этот знак свыше, и отметил его как спам. (Дайте танк) 

Пусть количество писем со спамом, которое пришло к нам на почту, имеет распределение Пуассона. Предположим, что вы получаете в среднем три спам-письма в день. Какова доля дней, в которые вы получаете пять или больше спам-писем?

```python
### ╰( ͡° ͜ʖ ͡° )つ▬▬ι═══════  bzzzzzzzzzz
# will the code be with you

p = sts.poisson(20)
```

```python
means = []
vars = []
for i in range(10, 1000):
    p = sts.poisson(i)
    poisson_vals = p.rvs(1000)
    means.append(poisson_vals.mean())
    vars.append(poisson_vals.var())
```

```python
import seaborn as sns

sns.regplot(x=means, y=vars, line_kws={'color':'blue'}, )
plt.plot([0, 1000], [0, 1000], c='g', linewidth=5, alpha=0.9)
plt.show()
```

```python
poisson_vals = p.rvs(1000)
```

```python
poisson_vals.mean(), poisson_vals.var()
```

```python
x = np.arange(1, 50)
cdf_ = p.cdf(x)
plt.scatter(x, cdf_)
```

```python
x = np.arange(1, 40)
cdf_ = p.pmf(x)
plt.scatter(x, cdf_)
plt.plot(x, cdf_)
```

```python
assert np.abs(p - 0.1847367) < 1e-5
```
